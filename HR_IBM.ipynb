<H1>Preprossing for dataset</H1>


1.   Convert categorize data to numerical data.
2.   devide dataset to X,Y.
3.   Normalize the dataset.



```python
df['Attrition'].value_counts(normalize=True)
```

```python
df['Attrition']= pd.get_dummies(df['Attrition'],drop_first=True)
df['Attrition']
```

```python
df['BusinessTravel'].value_counts(normalize=True)
```

```python
df['BusinessTravel'] = df['BusinessTravel'].replace({'Travel_Rarely':0,'Travel_Frequently':1,'Non-Travel':2})
df['BusinessTravel']
```

```python
df['Department'].value_counts(normalize=True)
```

```python
df['Department']= df['Department'].replace({'Research & Development':0,'Sales':1,'Human Resources':2})
df['Department']
```

```python
df['EducationField'].value_counts(normalize=True)
```

```python
df['EducationField']= df['EducationField'].replace({'Life Sciences':0,'Medical':1,'Marketing':2,'Technical Degree':3,'Other':4,'Human Resources':5})
df['EducationField']
```

```python
df['Gender'].value_counts(normalize=True)
```

```python
df['Gender']= pd.get_dummies(df['Gender'],drop_first=True)
```

```python
df['JobRole'].value_counts(normalize=True)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
enc= LabelEncoder()
```

```python
df['JobRole'] = enc.fit_transform(df['JobRole'])
df['JobRole']
```

```python
df['MaritalStatus'].value_counts(normalize=True)
```

```python
df['MaritalStatus']= df['MaritalStatus'].replace({'Married':0,'Single':1,'Divorced':2})
```

```python
df['OverTime'].value_counts(normalize=True)
```

```python
df['OverTime'] = pd.get_dummies(df['OverTime'],drop_first=True)
```

```python
df['Over18'].value_counts()
df['Over18'] = df['Over18'].replace({'Y':1})
```

```python
X = df.drop('Attrition',axis=1)
X
```

```python
y= df['Attrition']
y
```

```python
y= y.astype(int)
```

```python
y.dtype
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
ms = MinMaxScaler()
```

```python
X_train= ms.fit_transform(X_train)
```

```python
X_test= ms.fit_transform(X_test)
```

<h2> Artificial Neural Network </h2>


<h3> Model Building </h3>

```python
import tensorflow as tf
```

```python
from keras.models import Sequential
from keras.layers import Dense,Dropout,Activation
```

```python
ann = Sequential()
ann.add(Dense(100,input_dim = 34, activation = 'relu'))
ann.add(Dropout(0.5))
ann.add(Dense(50, activation='relu'))
ann.add(Dropout(0.5))
ann.add(Dense(50, activation='relu'))
ann.add(Dropout(0.5))
ann.add(Dense(50, activation='relu'))
ann.add(Dropout(0.5))
ann.add(Dense(50, activation='relu'))
ann.add(Dropout(0.5))
ann.add(Dense(1, activation='elu'))
ann.summary()
```

```python
ann.compile(optimizer='adam', loss='binary_crossentropy' , metrics=['accuracy'])
```

```python
ann.fit(X_train, y_train, batch_size=50, epochs=100 , verbose='auto', validation_split=0.3)
```

```python
ann.history.history
```
